```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from torch.autograd import Variable
from sklearn.model_selection import train_test_split
from tqdm import tqdm
import glob
from scipy import signal
import re

# ƒê∆∞·ªùng d·∫´n ƒë·∫øn th∆∞ m·ª•c d·ªØ li·ªáu BIDMC
BIDMC_PATH = '/content/drive/MyDrive/Iot-Mid/bidmc/bidmc_csv'
# ƒê∆∞·ªùng d·∫´n l∆∞u d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω
PROCESSED_PATH = '/content/drive/MyDrive/Iot-Mid/bidmc/processed'
# ƒê∆∞·ªùng d·∫´n l∆∞u m√¥ h√¨nh
MODEL_SAVE_PATH = '/content/drive/MyDrive/Iot-Mid/bidmc/models'
# ƒê∆∞·ªùng d·∫´n l∆∞u k·∫øt qu·∫£
RESULTS_PATH = '/content/drive/MyDrive/Iot-Mid/bidmc/results'

# T·∫°o th∆∞ m·ª•c n·∫øu ch∆∞a t·ªìn t·∫°i
for path in [PROCESSED_PATH, MODEL_SAVE_PATH, RESULTS_PATH]:
    os.makedirs(path, exist_ok=True)

# Thi·∫øt l·∫≠p seed ƒë·ªÉ k·∫øt qu·∫£ c√≥ th·ªÉ t√°i t·∫°o
torch.manual_seed(42)
np.random.seed(42)

# Ki·ªÉm tra v√† s·ª≠ d·ª•ng GPU n·∫øu c√≥
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(f"S·ª≠ d·ª•ng thi·∫øt b·ªã: {device}")

# ===================== PH·∫¶N X·ª¨ L√ù D·ªÆ LI·ªÜU =====================

def list_bidmc_files():
    """
    Li·ªát k√™ c√°c file trong b·ªô d·ªØ li·ªáu BIDMC v√† nh√≥m ch√∫ng theo ID b·ªánh nh√¢n
    """
    all_files = os.listdir(BIDMC_PATH)

    # T·∫°o t·ª´ ƒëi·ªÉn nh√≥m file theo ID b·ªánh nh√¢n
    patient_files = {}

    for file in all_files:
        # T√¨m ID b·ªánh nh√¢n t·ª´ t√™n file
        match = re.search(r'bidmc_(\d+)', file)
        if match:
            patient_id = str(int(match.group(1)))  # Lo·∫°i b·ªè s·ªë 0 ·ªü ƒë·∫ßu
            if patient_id not in patient_files:
                patient_files[patient_id] = {'signals': None, 'numerics': None, 'breaths': None}

            if 'Signal' in file:
                patient_files[patient_id]['signals'] = file
            elif 'Numerics' in file:
                patient_files[patient_id]['numerics'] = file
            elif 'Breaths' in file:
                patient_files[patient_id]['breaths'] = file
    return patient_files

def load_patient_data(patient_id, patient_files):
    """
    T·∫£i d·ªØ li·ªáu cho m·ªôt b·ªánh nh√¢n c·ª• th·ªÉ
    """

    files = patient_files[patient_id]
    data = {}

    # T·∫£i file t√≠n hi·ªáu
    if files['signals']:
        signals_path = os.path.join(BIDMC_PATH, files['signals'])
        data['signals'] = pd.read_csv(signals_path)

    # T·∫£i file th√¥ng s·ªë
    if files['numerics']:
        numerics_path = os.path.join(BIDMC_PATH, files['numerics'])
        data['numerics'] = pd.read_csv(numerics_path)

    # T·∫£i file nh·ªãp th·ªü
    if files['breaths']:
        breaths_path = os.path.join(BIDMC_PATH, files['breaths'])
        data['breaths'] = pd.read_csv(breaths_path)

    return data

def apply_bandpass_filter(data, fs=125, lowcut=0.5, highcut=8.0, order=4):
    """
    √Åp d·ª•ng b·ªô l·ªçc Butterworth bƒÉng th√¥ng cho t√≠n hi·ªáu
    """
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = signal.butter(order, [low, high], btype='band')
    return signal.filtfilt(b, a, data)

def normalize_signal(data):
    """
    Chu·∫©n h√≥a t√≠n hi·ªáu v·ªÅ kho·∫£ng 0-1
    """
    min_val = np.min(data)
    max_val = np.max(data)

    # Ki·ªÉm tra ƒë·ªÉ tr√°nh chia cho 0
    if max_val == min_val:
        return np.zeros_like(data)

    return (data - min_val) / (max_val - min_val)

def segment_data(data, fs=125, window_size=10):
    """
    Ph√¢n ƒëo·∫°n d·ªØ li·ªáu th√†nh c√°c c·ª≠a s·ªï c√≥ k√≠ch th∆∞·ªõc c·ªë ƒë·ªãnh
    """
    # S·ªë m·∫´u trong m·ªói c·ª≠a s·ªï
    samples_per_window = int(fs * window_size)

    # S·ªë c·ª≠a s·ªï ho√†n ch·ªânh c√≥ th·ªÉ t·∫°o ra
    num_windows = len(data) // samples_per_window

    # Ph√¢n ƒëo·∫°n d·ªØ li·ªáu
    segments = []
    for i in range(num_windows):
        start_idx = i * samples_per_window
        end_idx = start_idx + samples_per_window
        segment = data[start_idx:end_idx]
        segments.append(segment)

    return segments, [i * window_size for i in range(num_windows)]

def match_numerics_to_segment(segment_times, numerics_df):
    """
    L·∫•y gi√° tr·ªã HR v√† RR t·ª´ b·∫£ng numerics cho m·ªói ph√¢n ƒëo·∫°n
    """
    hr_values = []
    rr_values = []

    for start_time in segment_times:
        # T√¨m c√°c gi√° tr·ªã trong kho·∫£ng th·ªùi gian c·ªßa ph√¢n ƒëo·∫°n
        segment_numerics = numerics_df[(numerics_df['Time [s]'] >= start_time) &
                                       (numerics_df['Time [s]'] < start_time + 10)]

        if len(segment_numerics) > 0:
            # L·∫•y gi√° tr·ªã trung b√¨nh trong ƒëo·∫°n
            hr = segment_numerics[' HR'].mean()
            rr = segment_numerics[' RESP'].mean()
        else:
            # N·∫øu kh√¥ng c√≥ gi√° tr·ªã n√†o, l·∫•y gi√° tr·ªã g·∫ßn nh·∫•t
            closest_idx = (numerics_df['Time [s]'] - start_time).abs().idxmin()
            hr = numerics_df.loc[closest_idx, ' HR']
            rr = numerics_df.loc[closest_idx, ' RESP']
        hr_values.append(hr)
        rr_values.append(rr)

    return np.array(hr_values), np.array(rr_values)

def process_bidmc_data():
    """
    X·ª≠ l√Ω t·∫•t c·∫£ d·ªØ li·ªáu BIDMC v√† l∆∞u c√°c ph√¢n ƒëo·∫°n ƒë√£ x·ª≠ l√Ω
    """
    # Li·ªát k√™ t·∫•t c·∫£ c√°c file
    patient_files = list_bidmc_files()

    all_segments = {
        'ppg': [],  # PLETH signal
        'ecg': [],  # Lead II (ECG)
        'resp': [], # RESP signal
        'hr': [],   # Heart rate
        'rr': []    # Respiratory rate
    }

    for patient_id, files in patient_files.items():
        print(f"ƒêang x·ª≠ l√Ω d·ªØ li·ªáu b·ªánh nh√¢n {patient_id}...")

        # Ki·ªÉm tra xem c√≥ ƒë·ªß file kh√¥ng
        if not (files['signals'] and files['numerics']):
            print(f"Thi·∫øu file cho b·ªánh nh√¢n {patient_id}, b·ªè qua.")
            continue

        # T·∫£i d·ªØ li·ªáu
        data = load_patient_data(patient_id, patient_files)

        # X·ª≠ l√Ω t√≠n hi·ªáu
        # L·∫•y c√°c c·ªôt t√≠n hi·ªáu
        ppg = data['signals'][' PLETH'].values  # T√≠n hi·ªáu PPG
        ecg = data['signals'][' II'].values     # T√≠n hi·ªáu ECG (lead II)
        resp = data['signals'][' RESP'].values  # T√≠n hi·ªáu h√¥ h·∫•p

        # L·ªçc bƒÉng th√¥ng
        ppg_filtered = apply_bandpass_filter(ppg)
        ecg_filtered = apply_bandpass_filter(ecg)
        resp_filtered = apply_bandpass_filter(resp, lowcut=0.05, highcut=1.0)  # L·ªçc kh√°c cho h√¥ h·∫•p

        # Chu·∫©n h√≥a
        ppg_normalized = normalize_signal(ppg_filtered)
        ecg_normalized = normalize_signal(ecg_filtered)
        resp_normalized = normalize_signal(resp_filtered)

        # Ph√¢n ƒëo·∫°n
        ppg_segments, segment_times = segment_data(ppg_normalized)
        ecg_segments, _ = segment_data(ecg_normalized)
        resp_segments, _ = segment_data(resp_normalized)

        # L·∫•y th√¥ng s·ªë HR v√† RR cho m·ªói ph√¢n ƒëo·∫°n
        hr_values, rr_values = match_numerics_to_segment(segment_times, data['numerics'])

        # Th√™m v√†o danh s√°ch t·ªïng
        all_segments['ppg'].extend(ppg_segments)
        all_segments['ecg'].extend(ecg_segments)
        all_segments['resp'].extend(resp_segments)
        all_segments['hr'].extend(hr_values)
        all_segments['rr'].extend(rr_values)

        print(f"ƒê√£ x·ª≠ l√Ω {len(ppg_segments)} ph√¢n ƒëo·∫°n cho b·ªánh nh√¢n {patient_id}.")

    # Chuy·ªÉn ƒë·ªïi th√†nh m·∫£ng numpy
    for key in all_segments:
        all_segments[key] = np.array(all_segments[key])

    # L∆∞u d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω
    np.save(os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy'), all_segments)

    print(f"ƒê√£ ho√†n th√†nh x·ª≠ l√Ω {len(all_segments['ppg'])} ph√¢n ƒëo·∫°n t·ª´ t·∫•t c·∫£ b·ªánh nh√¢n.")

    # V·∫Ω m·ªôt v√†i v√≠ d·ª• ƒë·ªÉ ki·ªÉm tra
    visualize_processed_data(all_segments)

    return all_segments

def visualize_processed_data(segments, num_examples=3):
    """
    Tr·ª±c quan h√≥a m·ªôt s·ªë ph√¢n ƒëo·∫°n ƒë√£ x·ª≠ l√Ω
    """
    # Ch·ªçn ng·∫´u nhi√™n m·ªôt s·ªë ph√¢n ƒëo·∫°n
    indices = np.random.choice(len(segments['ppg']), num_examples, replace=False)

    fig, axs = plt.subplots(num_examples, 3, figsize=(15, 5 * num_examples))

    for i, idx in enumerate(indices):
        # V·∫Ω PPG
        axs[i, 0].plot(segments['ppg'][idx])
        axs[i, 0].set_title(f'PPG - Ph√¢n ƒëo·∫°n {idx}')

        # V·∫Ω ECG
        axs[i, 1].plot(segments['ecg'][idx])
        axs[i, 1].set_title(f'ECG - Ph√¢n ƒëo·∫°n {idx}')

        # V·∫Ω RESP
        axs[i, 2].plot(segments['resp'][idx])
        axs[i, 2].set_title(f'RESP - Ph√¢n ƒëo·∫°n {idx}')

        # Th√™m th√¥ng tin HR v√† RR
        axs[i, 0].set_xlabel(f'HR: {segments["hr"][idx]:.1f} BPM, RR: {segments["rr"][idx]:.1f} BrPM')

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'processed_data_examples.png'))
    plt.close()

# ===================== PH·∫¶N ƒê·ªäNH NGHƒ®A DATASET =====================

class BIDMCDatasetWithHRBR(Dataset):
    """
    Dataset cho d·ªØ li·ªáu BIDMC v·ªõi th√¥ng s·ªë HR v√† BR
    """
    def __init__(self, data_path=PROCESSED_PATH, filename='bidmc_processed_segments.npy'):
        """
        Kh·ªüi t·∫°o dataset
        """
        # T·∫£i d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω
        file_path = os.path.join(data_path, filename)
        self.segments = np.load(file_path, allow_pickle=True).item()

        # L·∫•y s·ªë l∆∞·ª£ng ph√¢n ƒëo·∫°n
        self.num_segments = len(self.segments['ppg'])

        # Chu·∫©n h√≥a HR v√† BR
        print("rr : ", self.segments['rr'])
        print("C√≥ gi√° tr·ªã NaN trong rr kh√¥ng?", np.isnan(self.segments['rr']).any())

        # Thay th·∫ø gi√° tr·ªã NaN b·∫±ng trung v·ªã c·ªßa rr
        self.segments['rr'] = np.nan_to_num(self.segments['rr'], nan=np.nanmedian(self.segments['rr']))

        # Thay th·∫ø gi√° tr·ªã 0 b·∫±ng trung v·ªã c·ªßa rr (sau khi ƒë√£ x·ª≠ l√Ω NaN)
        median_rr = np.median(self.segments['rr'][self.segments['rr'] > 0])  # Trung v·ªã c·ªßa c√°c gi√° tr·ªã > 0
        self.segments['rr'][self.segments['rr'] == 0] = median_rr

        print("C√≥ gi√° tr·ªã NaN trong rr kh√¥ng?", np.isnan(self.segments['rr']).any())
        print("rr sau khi thay th·∫ø 0 b·∫±ng trung v·ªã:", self.segments['rr'])
        self.hr_mean, self.hr_std = np.mean(self.segments['hr']), np.std(self.segments['hr'])
        self.rr_mean, self.rr_std = np.mean(self.segments['rr']), np.std(self.segments['rr'])

        self.normalized_hr = self.segments['hr']
        self.normalized_rr = self.segments['rr']

        print(f"Dataset ƒë∆∞·ª£c kh·ªüi t·∫°o v·ªõi {self.num_segments} ph√¢n ƒëo·∫°n")
        print(f"Ph·∫°m vi HR: {np.min(self.segments['hr']):.1f} - {np.max(self.segments['hr']):.1f} BPM, Trung b√¨nh: {self.hr_mean:.1f}, ƒê·ªô l·ªách chu·∫©n: {self.hr_std:.1f}")
        print(f"Ph·∫°m vi RR: {np.min(self.segments['rr']):.1f} - {np.max(self.segments['rr']):.1f} BrPM, Trung b√¨nh: {self.rr_mean:.1f}, ƒê·ªô l·ªách chu·∫©n: {self.rr_std:.1f}")

    def __len__(self):
        return self.num_segments

    def __getitem__(self, idx):
        # L·∫•y c√°c ph√¢n ƒëo·∫°n
        ppg = self.segments['ppg'][idx]
        ecg = self.segments['ecg'][idx]
        resp = self.segments['resp'][idx]
        hr = self.normalized_hr[idx]
        rr = self.normalized_rr[idx]

        # Chuy·ªÉn th√†nh tensor
        ppg_tensor = torch.FloatTensor(ppg)
        ecg_tensor = torch.FloatTensor(ecg)
        resp_tensor = torch.FloatTensor(resp)
        hr_tensor = torch.FloatTensor([hr])
        rr_tensor = torch.FloatTensor([rr])

        return ecg_tensor, ppg_tensor, resp_tensor, hr_tensor, rr_tensor

    def denormalize_params(self, hr, rr):
        """
        Chuy·ªÉn ƒë·ªïi HR v√† RR t·ª´ d·∫°ng chu·∫©n h√≥a v·ªÅ d·∫°ng g·ªëc
        """
        hr_denorm = hr * self.hr_std + self.hr_mean
        rr_denorm = rr * self.rr_std + self.rr_mean
        return hr_denorm, rr_denorm

def prepare_dataloaders(batch_size=32, test_size=0.2):
    """
    Chu·∫©n b·ªã DataLoader cho t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm th·ª≠
    """
    # Ki·ªÉm tra xem file d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω c√≥ t·ªìn t·∫°i kh√¥ng
    file_path = os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy')
    if not os.path.exists(file_path):
        print("D·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω kh√¥ng t·ªìn t·∫°i. ƒêang x·ª≠ l√Ω d·ªØ li·ªáu...")
        process_bidmc_data()

    # T·∫°o dataset
    dataset = BIDMCDatasetWithHRBR()

    # Chia t·∫≠p hu·∫•n luy·ªán v√† ki·ªÉm th·ª≠
    train_indices, val_indices = train_test_split(
        range(len(dataset)),
        test_size=test_size,
        random_state=42
    )

    # T·∫°o DataLoader
    train_loader = DataLoader(
        dataset,
        batch_size=batch_size,
        sampler=torch.utils.data.SubsetRandomSampler(train_indices),
        num_workers=2,
        pin_memory=True
    )

    val_loader = DataLoader(
        dataset,
        batch_size=batch_size,
        sampler=torch.utils.data.SubsetRandomSampler(val_indices),
        num_workers=2,
        pin_memory=True
    )

    return train_loader, val_loader, dataset

# ===================== PH·∫¶N ƒê·ªäNH NGHƒ®A M√î H√åNH =====================

class EncoderWithParams(nn.Module):
    """
    Encoder cho m√¥ h√¨nh cVAE v·ªõi th√¥ng s·ªë HR v√† BR
    """
    def __init__(self, input_dim=1250, hidden_dim=512, latent_dim=128, condition_dim=1250):
        super(EncoderWithParams, self).__init__()

        self.input_dim = input_dim
        self.hidden_dim = hidden_dim
        self.latent_dim = latent_dim

        # M·∫°ng Neural cho input signal (PPG)
        self.fc1 = nn.Linear(input_dim, hidden_dim)
        self.bn1 = nn.BatchNorm1d(hidden_dim)

        # M·∫°ng Neural cho conditional signal (ECG)
        self.cond_fc = nn.Linear(condition_dim, hidden_dim)
        self.cond_bn = nn.BatchNorm1d(hidden_dim)

        # M·∫°ng Neural cho t√≠n hi·ªáu h√¥ h·∫•p
        self.resp_fc = nn.Linear(condition_dim, hidden_dim // 2)
        self.resp_bn = nn.BatchNorm1d(hidden_dim // 2)

        # M·∫°ng Neural cho th√¥ng s·ªë HR v√† BR
        self.param_fc = nn.Linear(2, hidden_dim // 4)
        self.param_bn = nn.BatchNorm1d(hidden_dim // 4)

        # M·∫°ng k·∫øt h·ª£p
        combined_dim = hidden_dim * 2 + hidden_dim // 2 + hidden_dim // 4
        self.fc2 = nn.Linear(combined_dim, hidden_dim)
        self.bn2 = nn.BatchNorm1d(hidden_dim)

        # L·ªõp cho mean v√† variance
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_var = nn.Linear(hidden_dim, latent_dim)

    def forward(self, x, c, r, hr, br):
        # K·∫øt h·ª£p HR v√† BR
        params = torch.cat([hr, br], dim=1)

        # X·ª≠ l√Ω input signal (PPG)
        x = F.relu(self.bn1(self.fc1(x)))

        # X·ª≠ l√Ω conditional signal (ECG)
        c = F.relu(self.cond_bn(self.cond_fc(c)))

        # X·ª≠ l√Ω t√≠n hi·ªáu h√¥ h·∫•p
        r = F.relu(self.resp_bn(self.resp_fc(r)))

        # X·ª≠ l√Ω th√¥ng s·ªë
        p = F.relu(self.param_bn(self.param_fc(params)))

        # K·∫øt h·ª£p ƒë·∫∑c tr∆∞ng
        h = torch.cat([x, c, r, p], dim=1)
        h = F.relu(self.bn2(self.fc2(h)))

        # T√≠nh mean v√† variance
        mu = self.fc_mu(h)
        log_var = self.fc_var(h)

        return mu, log_var

class DecoderWithParams(nn.Module):
    """
    Decoder cho m√¥ h√¨nh cVAE v·ªõi th√¥ng s·ªë HR v√† BR
    """
    def __init__(self, latent_dim=128, hidden_dim=512, output_dim=1250, condition_dim=1250):
        super(DecoderWithParams, self).__init__()

        self.latent_dim = latent_dim
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim

        # M·∫°ng Neural cho latent vector
        self.fc1 = nn.Linear(latent_dim, hidden_dim)
        self.bn1 = nn.BatchNorm1d(hidden_dim)

        # M·∫°ng Neural cho conditional signal (ECG)
        self.cond_fc = nn.Linear(condition_dim, hidden_dim)
        self.cond_bn = nn.BatchNorm1d(hidden_dim)

        # M·∫°ng Neural cho t√≠n hi·ªáu h√¥ h·∫•p
        self.resp_fc = nn.Linear(condition_dim, hidden_dim // 2)
        self.resp_bn = nn.BatchNorm1d(hidden_dim // 2)

        # M·∫°ng Neural cho th√¥ng s·ªë HR v√† BR
        self.param_fc = nn.Linear(2, hidden_dim // 4)
        self.param_bn = nn.BatchNorm1d(hidden_dim // 4)

        # M·∫°ng k·∫øt h·ª£p
        combined_dim = hidden_dim * 2 + hidden_dim // 2 + hidden_dim // 4
        self.fc2 = nn.Linear(combined_dim, hidden_dim)
        self.bn2 = nn.BatchNorm1d(hidden_dim)

        # L·ªõp ƒë·∫ßu ra
        self.fc3 = nn.Linear(hidden_dim, output_dim)

    def forward(self, z, c, r, hr, br):
        # K·∫øt h·ª£p HR v√† BR
        params = torch.cat([hr, br], dim=1)

        # X·ª≠ l√Ω latent vector
        z = F.relu(self.bn1(self.fc1(z)))

        # X·ª≠ l√Ω conditional signal (ECG)
        c = F.relu(self.cond_bn(self.cond_fc(c)))

        # X·ª≠ l√Ω t√≠n hi·ªáu h√¥ h·∫•p
        r = F.relu(self.resp_bn(self.resp_fc(r)))

        # X·ª≠ l√Ω th√¥ng s·ªë
        p = F.relu(self.param_bn(self.param_fc(params)))

        # K·∫øt h·ª£p ƒë·∫∑c tr∆∞ng
        h = torch.cat([z, c, r, p], dim=1)
        h = F.relu(self.bn2(self.fc2(h)))

        # ƒê·∫ßu ra
        output = torch.sigmoid(self.fc3(h))  # Sigmoid do d·ªØ li·ªáu ƒë√£ ƒë∆∞·ª£c chu·∫©n h√≥a v·ªÅ 0-1

        return output

class ConditionalVAEWithParams(nn.Module):
    """
    M√¥ h√¨nh Conditional Variational Autoencoder (cVAE) v·ªõi th√¥ng s·ªë HR v√† BR
    """
    def __init__(self, input_dim=1250, hidden_dim=512, latent_dim=128, condition_dim=1250):
        super(ConditionalVAEWithParams, self).__init__()

        self.latent_dim = latent_dim

        # Encoder v√† Decoder
        self.encoder = EncoderWithParams(input_dim, hidden_dim, latent_dim, condition_dim)
        self.decoder = DecoderWithParams(latent_dim, hidden_dim, input_dim, condition_dim)

    def reparameterize(self, mu, log_var):
        """
        Reparameterization trick ƒë·ªÉ cho ph√©p backpropagation qua qu√° tr√¨nh l·∫•y m·∫´u
        """
        std = torch.exp(0.5 * log_var)
        eps = torch.randn_like(std)
        z = mu + eps * std
        return z

    def forward(self, x, c, r, hr, br):
        # Encode
        mu, log_var = self.encoder(x, c, r, hr, br)

        # Reparameterize
        z = self.reparameterize(mu, log_var)

        # Decode
        recon_x = self.decoder(z, c, r, hr, br)

        return recon_x, mu, log_var

    def generate(self, c, r, hr, br, num_samples=1):
        """
        Sinh t√≠n hi·ªáu t·ª´ ƒëi·ªÅu ki·ªán v√† th√¥ng s·ªë

        Tham s·ªë:
            c: T√≠n hi·ªáu ECG
            r: T√≠n hi·ªáu h√¥ h·∫•p
            hr: Nh·ªãp tim chu·∫©n h√≥a
            br: Nh·ªãp th·ªü chu·∫©n h√≥a
            num_samples: S·ªë l∆∞·ª£ng m·∫´u c·∫ßn sinh
        """
        # Chuy·ªÉn v·ªÅ thi·∫øt b·ªã ch√≠nh x√°c
        c = c.to(device)
        r = r.to(device)
        hr = hr.to(device)
        br = br.to(device)

        # T·∫°o latent vectors ng·∫´u nhi√™n
        z = torch.randn(num_samples, self.latent_dim).to(device)

        # Decode ƒë·ªÉ t·∫°o t√≠n hi·ªáu
        with torch.no_grad():
            # ƒê·∫£m b·∫£o hr v√† br c√≥ s·ªë chi·ªÅu ph√π h·ª£p
            if len(hr.shape) == 1:
                hr = hr.unsqueeze(1)
            if len(br.shape) == 1:
                br = br.unsqueeze(1)

            generated = self.decoder(z, c, r, hr, br)

        return generated

def vae_loss_function(recon_x, x, mu, log_var, beta=1.0):
    """
    H√†m m·∫•t m√°t cho VAE

    Tham s·ªë:
        recon_x: T√≠n hi·ªáu t√°i t·∫°o
        x: T√≠n hi·ªáu g·ªëc
        mu: Mean c·ªßa ph√¢n ph·ªëi latent
        log_var: Log variance c·ªßa ph√¢n ph·ªëi latent
        beta: Tr·ªçng s·ªë cho KL divergence

    Tr·∫£ v·ªÅ:
        loss: T·ªïng m·∫•t m√°t
        recon_loss: M·∫•t m√°t t√°i t·∫°o
        kl_loss: KL divergence
    """
    # Reconstruction loss (MSE)
    recon_loss = F.mse_loss(recon_x, x, reduction='sum')

    # KL divergence
    kl_loss = -0.5 * torch.sum(1 + log_var - mu.pow(2) - log_var.exp())

    # T·ªïng m·∫•t m√°t
    loss = recon_loss + beta * kl_loss

    return loss, recon_loss, kl_loss

# ===================== PH·∫¶N HU·∫§N LUY·ªÜN M√î H√åNH =====================

def train_cvae_with_params(model, train_loader, val_loader, epochs=100, lr=1e-4, beta=1.0, device=device):
    """
    Hu·∫•n luy·ªán m√¥ h√¨nh cVAE v·ªõi th√¥ng s·ªë HR v√† BR
    """
    optimizer = optim.Adam(model.parameters(), lr=lr)
    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, 'min', patience=5, factor=0.5)

    # L∆∞u l·ªãch s·ª≠ m·∫•t m√°t
    train_losses = []
    train_recon_losses = []
    train_kl_losses = []
    val_losses = []
    val_recon_losses = []
    val_kl_losses = []

    best_val_loss = float('inf')

    for epoch in range(epochs):
        model.train()
        train_loss = 0
        train_recon_loss = 0
        train_kl_loss = 0
        for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(tqdm(train_loader, desc=f"Epoch {epoch+1}/{epochs}")):

            ecg = ecg.to(device)
            ppg = ppg.to(device)
            resp = resp.to(device)
            hr = hr.to(device)
            rr = rr.to(device)

            # L√†m ph·∫≥ng d·ªØ li·ªáu
            ecg_flat = ecg.view(ecg.size(0), -1)
            ppg_flat = ppg.view(ppg.size(0), -1)
            resp_flat = resp.view(resp.size(0), -1)

            optimizer.zero_grad()

            # Forward pass
            recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

            # T√≠nh m·∫•t m√°t
            loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

            # Backward pass
            loss.backward()
            optimizer.step()

            train_loss += loss.item()
            train_recon_loss += recon.item()
            train_kl_loss += kl.item()

        # ƒê√°nh gi√° tr√™n t·∫≠p validation
        model.eval()
        val_loss = 0
        val_recon_loss = 0
        val_kl_loss = 0

        with torch.no_grad():
            for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(val_loader):
                ecg = ecg.to(device)
                ppg = ppg.to(device)
                resp = resp.to(device)
                hr = hr.to(device)
                rr = rr.to(device)

                # L√†m ph·∫≥ng d·ªØ li·ªáu
                ecg_flat = ecg.view(ecg.size(0), -1)
                ppg_flat = ppg.view(ppg.size(0), -1)
                resp_flat = resp.view(resp.size(0), -1)

                # Forward pass
                recon_batch, mu, log_var = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

                # T√≠nh m·∫•t m√°t
                loss, recon, kl = vae_loss_function(recon_batch, ppg_flat, mu, log_var, beta)

                val_loss += loss.item()
                val_recon_loss += recon.item()
                val_kl_loss += kl.item()

        # T√≠nh trung b√¨nh m·∫•t m√°t
        train_loss /= len(train_loader)
        train_recon_loss /= len(train_loader)
        train_kl_loss /= len(train_loader)

        val_loss /= len(val_loader)
        val_recon_loss /= len(val_loader)
        val_kl_loss /= len(val_loader)


        # C·∫≠p nh·∫≠t scheduler
        scheduler.step(val_loss)

        # L∆∞u l·∫°i l·ªãch s·ª≠ m·∫•t m√°t
        train_losses.append(train_loss)
        train_recon_losses.append(train_recon_loss)
        train_kl_losses.append(train_kl_loss)
        val_losses.append(val_loss)
        val_recon_losses.append(val_recon_loss)
        val_kl_losses.append(val_kl_loss)

        # In m·∫•t m√°t
        print(f"Epoch {epoch+1}/{epochs}")
        print(f"Train Loss: {train_loss:.4f} (Recon: {train_recon_loss:.4f}, KL: {train_kl_loss:.4f})")
        print(f"Val Loss: {val_loss:.4f} (Recon: {val_recon_loss:.4f}, KL: {val_kl_loss:.4f})")

        # L∆∞u m√¥ h√¨nh t·ªët nh·∫•t
        if val_loss < best_val_loss:
            best_val_loss = val_loss
            torch.save(model.state_dict(), os.path.join(MODEL_SAVE_PATH, 'best_cvae_with_params.pth'))
            print("ƒê√£ l∆∞u m√¥ h√¨nh t·ªët nh·∫•t!")

        # L∆∞u m√¥ h√¨nh cu·ªëi c√πng
        torch.save(model.state_dict(), os.path.join(MODEL_SAVE_PATH, 'last_cvae_with_params.pth'))

        # Tr·ª±c quan h√≥a k·∫øt qu·∫£ m·ªói 10 epoch
        if (epoch + 1) % 10 == 0 or epoch == 0:
            visualize_results(model, val_loader, epoch + 1, dataset=val_loader.dataset)

    # V·∫Ω ƒë·ªì th·ªã m·∫•t m√°t
    plot_loss_curves(train_losses, val_losses,
                     train_recon_losses, val_recon_losses,
                     train_kl_losses, val_kl_losses)

    return model, (train_losses, val_losses)
def plot_ppg_spectrum(original_ppg, reconstructed_ppg, fs=125, title="Ph·ªï t√≠n hi·ªáu PPG tr∆∞·ªõc v√† sau khi qua m√¥ h√¨nh"):
    """
    V·∫Ω ph·ªï t√≠n hi·ªáu PPG g·ªëc v√† PPG ƒë√£ t√°i t·∫°o b·∫±ng m√¥ h√¨nh.

    Parameters:
    - original_ppg: T√≠n hi·ªáu PPG ban ƒë·∫ßu (tr∆∞·ªõc khi v√†o model)
    - reconstructed_ppg: T√≠n hi·ªáu PPG ƒë∆∞·ª£c t√°i t·∫°o t·ª´ model
    - fs: T·∫ßn s·ªë l·∫•y m·∫´u (m·∫∑c ƒë·ªãnh 125Hz)
    """
    # T√≠nh FFT
    original_fft = np.abs(np.fft.fft(original_ppg))
    reconstructed_fft = np.abs(np.fft.fft(reconstructed_ppg))

    # Tr·ª•c t·∫ßn s·ªë
    freqs = np.fft.fftfreq(len(original_ppg), d=1/fs)

    # L·ªçc ch·ªâ l·∫•y ph·∫ßn t·∫ßn s·ªë d∆∞∆°ng
    mask = freqs > 0
    freqs = freqs[mask]
    original_fft = original_fft[mask]
    reconstructed_fft = reconstructed_fft[mask]

    # V·∫Ω bi·ªÉu ƒë·ªì ph·ªï t√≠n hi·ªáu
    plt.figure(figsize=(10, 5))
    plt.plot(freqs, original_fft, label="PPG G·ªëc", alpha=0.7)
    plt.plot(freqs, reconstructed_fft, label="PPG T√°i t·∫°o", alpha=0.7, linestyle='dashed')
    plt.xlabel("T·∫ßn s·ªë (Hz)")
    plt.ylabel("Bi√™n ƒë·ªô")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.show()

def compare_ppg_spectrum(model, val_loader, device):
    """
    So s√°nh ph·ªï PPG tr∆∞·ªõc v√† sau khi qua m√¥ h√¨nh.
    """
    model.eval()  # Chuy·ªÉn m√¥ h√¨nh sang ch·∫ø ƒë·ªô ƒë√°nh gi√°
    with torch.no_grad():
        for batch_idx, (ecg, ppg, resp, hr, rr) in enumerate(val_loader):
            # Ch·ªâ l·∫•y m·ªôt batch ƒë·∫ßu ti√™n
            ecg = ecg.to(device)
            ppg = ppg.to(device)
            resp = resp.to(device)
            hr = hr.to(device)
            rr = rr.to(device)

            # L√†m ph·∫≥ng d·ªØ li·ªáu
            ecg_flat = ecg.view(ecg.size(0), -1)
            ppg_flat = ppg.view(ppg.size(0), -1)
            resp_flat = resp.view(resp.size(0), -1)

            # üõ† In d·ªØ li·ªáu ƒë·∫ßu v√†o
            print(f"Batch {batch_idx}:")
            print(f"  ECG shape: {ecg.shape}, Flattened: {ecg_flat.shape}")
            print(f"  PPG shape: {ppg.shape}, Flattened: {ppg_flat.shape}")
            print(f"  Resp shape: {resp.shape}, Flattened: {resp_flat.shape}")
            print(f"  HR: {hr.cpu().numpy()}, RR: {rr.cpu().numpy()}")

            # Forward qua m√¥ h√¨nh
            recon_ppg, _, _ = model(ppg_flat, ecg_flat, resp_flat, hr, rr)

            # Chuy·ªÉn v·ªÅ numpy ƒë·ªÉ v·∫Ω ph·ªï
            original_ppg = ppg_flat.cpu().numpy()[0]  # L·∫•y t√≠n hi·ªáu ƒë·∫ßu ti√™n
            reconstructed_ppg = recon_ppg.cpu().numpy()[0]  # L·∫•y t√≠n hi·ªáu ƒë·∫ßu ti√™n

            # üõ† In d·ªØ li·ªáu sau khi qua model
            print(f"  Original PPG sample (first 5 values): {original_ppg}")
            print(f"  Reconstructed PPG sample (first 5 values): {reconstructed_ppg}")

            # V·∫Ω ph·ªï t√≠n hi·ªáu
            plot_ppg_spectrum(original_ppg, reconstructed_ppg)
            break  # Ch·ªâ l·∫•y m·ªôt batch


def plot_loss_curves(train_losses, val_losses, train_recon_losses, val_recon_losses, train_kl_losses, val_kl_losses):
    """
    V·∫Ω ƒë·ªì th·ªã m·∫•t m√°t
    """
    plt.figure(figsize=(15, 5))

    # ƒê·ªì th·ªã t·ªïng m·∫•t m√°t
    plt.subplot(1, 3, 1)
    plt.plot(train_losses, label='Train')
    plt.plot(val_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Total Loss')
    plt.legend()

    # ƒê·ªì th·ªã reconstruction loss
    plt.subplot(1, 3, 2)
    plt.plot(train_recon_losses, label='Train')
    plt.plot(val_recon_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Reconstruction Loss')
    plt.legend()

    # ƒê·ªì th·ªã KL loss
    plt.subplot(1, 3, 3)
    plt.plot(train_kl_losses, label='Train')
    plt.plot(val_kl_losses, label='Validation')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('KL Divergence Loss')
    plt.legend()

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'cvae_loss_curves.png'))
    plt.close()

def visualize_results(model, data_loader, epoch, num_samples=3, dataset=None):
    """
    Tr·ª±c quan h√≥a k·∫øt qu·∫£ c·ªßa m√¥ h√¨nh v·ªõi c√°c th√¥ng s·ªë HR v√† BR kh√°c nhau
    """
    model.eval()

    # L·∫•y m·ªôt batch d·ªØ li·ªáu
    ecg, ppg, resp, hr, rr = next(iter(data_loader))

    # Ch·ªçn m·ªôt s·ªë m·∫´u
    ecg = ecg[:num_samples].to(device)
    ppg = ppg[:num_samples].to(device)
    resp = resp[:num_samples].to(device)
    hr_orig = hr[:num_samples].to(device)
    rr_orig = rr[:num_samples].to(device)

    # L√†m ph·∫≥ng d·ªØ li·ªáu
    ecg_flat = ecg.view(ecg.size(0), -1)
    ppg_flat = ppg.view(ppg.size(0), -1)
    resp_flat = resp.view(resp.size(0), -1)

    # T·∫°o c√°c gi√° tr·ªã HR v√† BR kh√°c nhau ƒë·ªÉ ki·ªÉm tra
    hr_values = [-1.0, 0.0, 1.0]  # Gi√° tr·ªã chu·∫©n h√≥a (th·∫•p, trung b√¨nh, cao)
    rr_values = [-1.0, 0.0, 1.0]  # Gi√° tr·ªã chu·∫©n h√≥a (th·∫•p, trung b√¨nh, cao)

    # Gi·∫£i m√£ gi√° tr·ªã HR v√† RR th·ª±c t·∫ø
    hr_actual = []
    rr_actual = []

    for i in range(num_samples):
        if dataset:
            hr_val, rr_val = dataset.denormalize_params(hr_orig[i].item(), rr_orig[i].item())
            hr_actual.append(hr_val)
            rr_actual.append(rr_val)
        else:
            hr_actual.append(f"Norm: {hr_orig[i].item():.2f}")
            rr_actual.append(f"Norm: {rr_orig[i].item():.2f}")

    # Gi·∫£i m√£ c√°c gi√° tr·ªã HR v√† RR ƒë∆∞·ª£c s·ª≠ d·ª•ng trong th·ª≠ nghi·ªám
    hr_denorm_values = []
    rr_denorm_values = []

    for hr_val in hr_values:
        for rr_val in rr_values:
            if dataset:
                hr_denorm, rr_denorm = dataset.denormalize_params(hr_val, rr_val)
                hr_denorm_values.append(hr_denorm)
                rr_denorm_values.append(rr_denorm)
            else:
                hr_denorm_values.append(f"Norm: {hr_val:.2f}")
                rr_denorm_values.append(f"Norm: {rr_val:.2f}")

    # T·∫°o grid ƒë·ªÉ hi·ªÉn th·ªã
    fig, axs = plt.subplots(num_samples, len(hr_values) * len(rr_values) + 2,
                          figsize=(4 * (len(hr_values) * len(rr_values) + 2), 3 * num_samples))

    for i in range(num_samples):
        # Hi·ªÉn th·ªã t√≠n hi·ªáu ECG
        axs[i, 0].plot(ecg_flat[i].cpu().numpy())
        axs[i, 0].set_title(f'ECG\nHR={hr_actual[i]}\nRR={rr_actual[i]}')
        axs[i, 0].set_yticks([])

        # Hi·ªÉn th·ªã t√≠n hi·ªáu PPG th·ª±c
        axs[i, 1].plot(ppg_flat[i].cpu().numpy())
        axs[i, 1].set_title('PPG Th·ª±c')
        axs[i, 1].set_yticks([])

        # T·∫°o v√† hi·ªÉn th·ªã t√≠n hi·ªáu v·ªõi c√°c th√¥ng s·ªë kh√°c nhau
        col_idx = 2
        for hr_idx, hr_val in enumerate(hr_values):
            for rr_idx, rr_val in enumerate(rr_values):
                # T·∫°o tensor cho HR v√† RR
                hr_tensor = torch.FloatTensor([[hr_val]]).to(device)
                rr_tensor = torch.FloatTensor([[rr_val]]).to(device)

                # Sinh t√≠n hi·ªáu m·ªõi
                generated = model.generate(
                    ecg_flat[i].unsqueeze(0),
                    resp_flat[i].unsqueeze(0),
                    hr_tensor,
                    rr_tensor
                )

                # Hi·ªÉn th·ªã
                axs[i, col_idx].plot(generated[0].cpu().detach().numpy())
                axs[i, col_idx].set_title(f'PPG Sinh\nHR={hr_denorm_values[col_idx-2]}\nRR={rr_denorm_values[col_idx-2]}')
                axs[i, col_idx].set_yticks([])

                col_idx += 1

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, f'cvae_results_epoch_{epoch}.png'))
    plt.close()

def experiment_with_parameter_ranges(model, data_loader, dataset):
    """
    Th√≠ nghi·ªám v·ªõi nhi·ªÅu gi√° tr·ªã HR v√† BR kh√°c nhau
    """
    # L·∫•y m·ªôt m·∫´u d·ªØ li·ªáu
    ecg, ppg, resp, _, _ = next(iter(data_loader))
    ecg = ecg[0].view(1, -1).to(device)
    resp = resp[0].view(1, -1).to(device)

    # T·∫°o c√°c gi√° tr·ªã HR v√† BR ƒë·ªÉ th·ª≠ nghi·ªám
    hr_values = np.linspace(50, 120, 5)  # T·ª´ 50 ƒë·∫øn 120 BPM
    rr_values = np.linspace(8, 30, 5)    # T·ª´ 8 ƒë·∫øn 30 BrPM

        # Chu·∫©n h√≥a c√°c gi√° tr·ªã
    hr_norm = hr_values
    rr_norm = rr_values
    # T·∫°o grid ƒë·ªÉ hi·ªÉn th·ªã
    fig, axs = plt.subplots(len(hr_norm), len(rr_norm), figsize=(15, 15))

    for i, hr in enumerate(hr_norm):
        for j, rr in enumerate(rr_norm):
            # Sinh t√≠n hi·ªáu
            hr_tensor = torch.FloatTensor([[hr]]).to(device)
            rr_tensor = torch.FloatTensor([[rr]]).to(device)

            generated = model.generate(ecg, resp, hr_tensor, rr_tensor)

            # Hi·ªÉn th·ªã
            axs[i, j].plot(generated[0].cpu().detach().numpy())
            if dataset:
                axs[i, j].set_title(f'HR={hr_values[i]:.1f} BPM, RR={rr_values[j]:.1f} BrPM')
            else:
                axs[i, j].set_title(f'HR norm={hr:.2f}, RR norm={rr:.2f}')
            axs[i, j].set_xticks([])
            axs[i, j].set_yticks([])

    plt.tight_layout()
    plt.savefig(os.path.join(RESULTS_PATH, 'parameter_range_experiment.png'))
    plt.close()

# ===================== PH·∫¶N CH·∫†Y TH·ª¨ NGHI·ªÜM =====================

def run_experiment(epochs=30):
    """
    Ch·∫°y th·ª≠ nghi·ªám m√¥ h√¨nh cVAE v·ªõi th√¥ng s·ªë HR v√† BR
    """
    # Th√¥ng s·ªë
    batch_size = 32
    test_size = 0.2
    input_dim = 1250  # K√≠ch th∆∞·ªõc c·ª≠a s·ªï 10s v·ªõi t·∫ßn s·ªë l·∫•y m·∫´u 125Hz
    latent_dim = 128
    hidden_dim = 512

    # Chu·∫©n b·ªã d·ªØ li·ªáu
    train_loader, val_loader, dataset = prepare_dataloaders(batch_size, test_size)

    # Kh·ªüi t·∫°o m√¥ h√¨nh
    print("\n=== Kh·ªüi t·∫°o m√¥ h√¨nh Conditional VAE v·ªõi th√¥ng s·ªë HR v√† BR ===")
    cvae_params = ConditionalVAEWithParams(input_dim, hidden_dim, latent_dim, input_dim).to(device)

    # Hu·∫•n luy·ªán m√¥ h√¨nh
    print("\n=== B·∫Øt ƒë·∫ßu hu·∫•n luy·ªán ===")
    model, losses = train_cvae_with_params(cvae_params, train_loader, val_loader, epochs=epochs, lr=1e-4, beta=0.1, device=device)
    print("\n=== So s√°nh ph·ªï t√≠n hi·ªáu PPG tr∆∞·ªõc v√† sau khi qua m√¥ h√¨nh ===")
    compare_ppg_spectrum(model, val_loader, device)

    # Th·ª≠ nghi·ªám v·ªõi c√°c th√¥ng s·ªë kh√°c nhau
    print("\n=== Th√≠ nghi·ªám v·ªõi c√°c th√¥ng s·ªë HR v√† BR kh√°c nhau ===")
    experiment_with_parameter_ranges(model, val_loader, dataset)

    print("\n=== Ho√†n th√†nh ===")
    return model, dataset

if __name__ == "__main__":
    # X·ª≠ l√Ω d·ªØ li·ªáu n·∫øu ch∆∞a c√≥
    if not os.path.exists(os.path.join(PROCESSED_PATH, 'bidmc_processed_segments.npy')):
        print("B·∫Øt ƒë·∫ßu x·ª≠ l√Ω d·ªØ li·ªáu BIDMC...")
        process_bidmc_data()

    # Ch·∫°y th·ª≠ nghi·ªám
    print("B·∫Øt ƒë·∫ßu hu·∫•n luy·ªán m√¥ h√¨nh cVAE...")
    model, dataset = run_experiment(epochs=30)
```
